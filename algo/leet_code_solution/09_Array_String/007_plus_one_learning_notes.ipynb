## Problem Description

You are given a **large integer** represented as an array `digits`, where each `digits[i]` is the `i`th digit of the integer.

Increment the large integer by one and return the resulting array.

### Examples:

**Example 1:**
```
Input: digits = [1,2,3]
Output: [1,2,4]
Explanation: 123 + 1 = 124
```

**Example 2:**
```
Input: digits = [9]
Output: [1,0]
Explanation: 9 + 1 = 10
```

**Example 3:**
```
Input: digits = [9,9,9]
Output: [1,0,0,0]
Explanation: 999 + 1 = 1000
```

## Key Insights

### Strategy:

1. Start from **rightmost digit** (least significant)
2. Add 1
3. If **< 10** ‚Üí Done!
4. If **= 10** ‚Üí Set to 0, carry to next
5. If all 9s ‚Üí Prepend 1

## Solution

````python
from typing import List

def plusOne(digits: List[int]) -> List[int]:
    """
    Time: O(n)
    Space: O(1) or O(n) if all 9s
    """
    n = len(digits)
    
    # Iterate from right to left
    for i in range(n - 1, -1, -1):
        # If not 9, just add 1 and return
        if digits[i] < 9:
            digits[i] += 1
            return digits
        
        # If 9, set to 0 and continue (carry)
        digits[i] = 0
    
    # If we're here, all digits were 9
    # Result is 1 followed by all 0s
    return [1] + digits
````

## Test Cases

````python
print(plusOne([1,2,3]))  # [1,2,4]
print(plusOne([9]))  # [1,0]
print(plusOne([9,9,9]))  # [1,0,0,0]
print(plusOne([1,9,9]))  # [2,0,0]
print(plusOne([0]))  # [1]
````
