## Problem Description

Given an integer array `nums`, return all the **triplets** `[nums[i], nums[j], nums[k]]` such that:
- `i != j`, `i != k`, and `j != k`
- `nums[i] + nums[j] + nums[k] == 0`

**Notice:** The solution set must not contain duplicate triplets.

### Examples:

**Example 1:**
```
Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
```

**Example 2:**
```
Input: nums = [0,1,1]
Output: []
```

**Example 3:**
```
Input: nums = [0,0,0]
Output: [[0,0,0]]
```

## Key Insights

### The Strategy:

1. **Sort** the array
2. **Fix one element** (i)
3. Use **two pointers** (left, right) to find pair that sums to `-nums[i]`
4. **Skip duplicates** to avoid duplicate triplets

### Why Sort?
- Makes it easy to **skip duplicates**
- Enables **two-pointer** technique
- Move pointers based on sum comparison

## Solution

````python
from typing import List

def threeSum(nums: List[int]) -> List[List[int]]:
    """
    Time: O(n¬≤)
    Space: O(1) excluding output
    """
    nums.sort()  # O(n log n)
    result = []
    n = len(nums)
    
    for i in range(n - 2):
        # Skip duplicates for i
        if i > 0 and nums[i] == nums[i - 1]:
            continue
        
        # Two pointers
        left, right = i + 1, n - 1
        target = -nums[i]
        
        while left < right:
            current_sum = nums[left] + nums[right]
            
            if current_sum == target:
                result.append([nums[i], nums[left], nums[right]])
                
                # Skip duplicates for left
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                # Skip duplicates for right
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                
                left += 1
                right -= 1
            
            elif current_sum < target:
                left += 1
            else:
                right -= 1
    
    return result
````

## Test Cases

````python
print(threeSum([-1,0,1,2,-1,-4]))  # [[-1,-1,2],[-1,0,1]]
print(threeSum([0,1,1]))  # []
print(threeSum([0,0,0]))  # [[0,0,0]]
print(threeSum([]))  # []
````

### What I Learned:

2. **Skip duplicates** at all levels
3. Fix one, find two (reduces to Two Sum)
