## Problem Description

You are given an integer array `height` of length `n`. There are `n` vertical lines where the two endpoints of the `i`th line are `(i, 0)` and `(i, height[i])`.

Find two lines that together with the x-axis form a container, such that the container contains the **most water**.

Return the **maximum** amount of water a container can store.

### Examples:

**Example 1:**
```
Input: height = [1,8,6,2,5,4,8,3,7]
Output: 49
Explanation: Max area between lines at index 1 and 8 is 49.
```

**Example 2:**
```
Input: height = [1,1]
Output: 1
```

## Key Insights

### The Key Insight:

**Area = min(height[left], height[right]) √ó (right - left)**

### Strategy:

1. Start with **widest** container (left=0, right=n-1)
2. Calculate area
3. **Move the shorter line** inward (greedy choice)
   - Why? Moving taller line can't increase area
   - Moving shorter line might find taller partner
4. Track max area

## Solution

````python
from typing import List

def maxArea(height: List[int]) -> int:
    """
    Two Pointers Approach
    
    Time: O(n)
    Space: O(1)
    """
    left, right = 0, len(height) - 1
    max_area = 0
    
    while left < right:
        # Calculate current area
        width = right - left
        current_height = min(height[left], height[right])
        area = width * current_height
        
        # Update max
        max_area = max(max_area, area)
        
        # Move the shorter line
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
    
    return max_area
````

## Test Cases

````python
print(maxArea([1,8,6,2,5,4,8,3,7]))  # 49
print(maxArea([1,1]))  # 1
print(maxArea([4,3,2,1,4]))  # 16
print(maxArea([1,2,1]))  # 2
````

## Key Takeaways

### What I Learned:
1. **Greedy choice** - Move shorter line
2. **Two pointers** from both ends
3. Width decreases, need taller height

### Why Move Shorter Line?
- Area limited by shorter height
- Moving taller line ‚Üí width ‚Üì, height can't ‚Üë ‚Üí area ‚Üì
- Moving shorter line ‚Üí might find taller ‚Üí area might ‚Üë
