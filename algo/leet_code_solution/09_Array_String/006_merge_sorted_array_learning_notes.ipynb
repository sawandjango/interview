## Problem Description

You are given two integer arrays `nums1` and `nums2`, sorted in **non-decreasing order**, and two integers `m` and `n`, representing the number of elements in `nums1` and `nums2` respectively.

**Merge** `nums2` into `nums1` as one sorted array.

The final sorted array should be stored **inside `nums1`**. `nums1` has length `m + n` with the first `m` elements being the values to merge, and the last `n` elements set to 0 (to be ignored).

### Examples:

**Example 1:**
```
Input: nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3
Output: [1,2,2,3,5,6]
```

**Example 2:**
```
Input: nums1 = [1], m = 1, nums2 = [], n = 0
Output: [1]
```

## Key Insights

### The Key Insight:

**Fill from the END to avoid overwriting elements!**

‚ùå Wrong: Start from beginning (will overwrite nums1 elements)  
‚úÖ Right: Start from end (empty space available)

### Strategy:

Use **three pointers**:
1. `p1` ‚Üí Last element of nums1 (at m-1)
2. `p2` ‚Üí Last element of nums2 (at n-1)
3. `p` ‚Üí Last position to fill (at m+n-1)

## Solution - Three Pointers from End

````python
from typing import List

def merge(nums1: List[int], m: int, nums2: List[int], n: int) -> None:
    """
    Merge in-place from the end
    
    Time: O(m + n)
    Space: O(1)
    """
    # Pointers for nums1, nums2, and merged position
    p1 = m - 1  # Last element in nums1
    p2 = n - 1  # Last element in nums2
    p = m + n - 1  # Last position in merged array
    
    # Merge from the end
    while p2 >= 0:
        if p1 >= 0 and nums1[p1] > nums2[p2]:
            nums1[p] = nums1[p1]
            p1 -= 1
        else:
            nums1[p] = nums2[p2]
            p2 -= 1
        p -= 1
````

## Test Cases

````python
# Test Case 1
nums1 = [1,2,3,0,0,0]
merge(nums1, 3, [2,5,6], 3)
print(nums1)  # [1,2,2,3,5,6]

# Test Case 2
nums1 = [1]
merge(nums1, 1, [], 0)
print(nums1)  # [1]

# Test Case 3
nums1 = [0]
merge(nums1, 0, [1], 1)
print(nums1)  # [1]
````
