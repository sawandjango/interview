## Problem Description

A phrase is a **palindrome** if, after converting all uppercase letters into lowercase letters and removing all non-alphanumeric characters, it reads the same forward and backward.

Given a string `s`, return `true` if it is a palindrome, or `false` otherwise.

### Examples:

**Example 1:**
```
Input: s = "A man, a plan, a canal: Panama"
Output: true
Explanation: "amanaplanacanalpanama" is a palindrome.
```

**Example 2:**
```
Input: s = "race a car"
Output: false
Explanation: "raceacar" is not a palindrome.
```

**Example 3:**
```
Input: s = " "
Output: true
Explanation: Empty string is considered a palindrome.
```

## Key Insights

### The Key Insight:

**Use two pointers - one from start, one from end - compare while moving towards center!**

### Strategy:

1. **Ignore non-alphanumeric** characters
2. **Compare case-insensitively** (convert to lowercase)
3. **Two pointers** - left and right
4. **Move inward** - skip invalid chars, compare valid ones

## Approach & Strategy

### Complexity Analysis:

| Approach | Time | Space |
|----------|------|-------|
| **Two Pointers** | ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê O(n) | ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê O(1) |
| **Clean & Reverse** | ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê O(n) | ‚≠ê‚≠ê‚≠ê O(n) |

## Solution 1: Two Pointers (Optimal)

````python
def isPalindrome(s: str) -> bool:
    """
    Two Pointers Approach
    
    Time: O(n)
    Space: O(1)
    """
    left, right = 0, len(s) - 1
    
    while left < right:
        # Skip non-alphanumeric from left
        while left < right and not s[left].isalnum():
            left += 1
        
        # Skip non-alphanumeric from right
        while left < right and not s[right].isalnum():
            right -= 1
        
        # Compare characters (case-insensitive)
        if s[left].lower() != s[right].lower():
            return False
        
        left += 1
        right -= 1
    
    return True
````

## Solution 2: Clean and Reverse (Simpler)

````python
def isPalindrome_Clean(s: str) -> bool:
    """
    Clean string then compare with reverse
    
    Time: O(n)
    Space: O(n)
    """
    # Keep only alphanumeric and convert to lowercase
    cleaned = ''.join(c.lower() for c in s if c.isalnum())
    
    # Check if equals its reverse
    return cleaned == cleaned[::-1]
````

## Test Cases

````python
# Test Case 1: Standard palindrome
s1 = "A man, a plan, a canal: Panama"
print(f"'{s1}': {isPalindrome(s1)}")  # True

# Test Case 2: Not a palindrome
s2 = "race a car"
print(f"'{s2}': {isPalindrome(s2)}")  # False

# Test Case 3: Empty/space
s3 = " "
print(f"'{s3}': {isPalindrome(s3)}")  # True

# Test Case 4: Single character
s4 = "a"
print(f"'{s4}': {isPalindrome(s4)}")  # True

# Test Case 5: Numbers and letters
s5 = "0P"
print(f"'{s5}': {isPalindrome(s5)}")  # False
````
